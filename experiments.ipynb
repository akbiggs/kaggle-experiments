```python
net = nl.net.newff([[-0.5, 0.5]] * 93, [9])
```

```python
input = x.view(np.float64).reshape(x.shape + (-1,))
target = expected.view(np.int).reshape(expected.shape + (-1,))

def convert_class_to_flags(class_row):
    n = class_row[0]
    flags = [0] * 9
    flags[n - 1] = 1
    return flags

target = np.array(map(convert_class_to_flags, target))
```

```python
error = net.train(input, target, epochs=500, show=100, goal=0.02)
error
```
